**Context:**
Top 50 most streamed tracks on Spotify in 2020. This dataset has various variables regarding these songs.

**String Variables:**
- artist: name of the artist
- album: name of the album the song is on
- track_name: name of the song
- track_id: ID of the song
- genre: genre of the song


**Numerical Variables:**
- ID: index column
- key: primary identifier of songs
- energy: higher value the more energetic the song (range: 0 to 1)
- danceability: higher value the more danceable (range: 0 to 1)
- instrumentalness: reflects presence of vocal; the higher the less vocals (range: 0 to 1)
- valence: positiveness of the song; the higher the more positive (range: 0 to 1)
- acousticness: acousticness level of the song (range: 0 to 1)
- duration_ms( Generally has a range of 200k-400k)
- liveness: likeleness to have been recorded with live audience (range: 0 to 1)
- speechiness: reflects presence of spoken word in the song (range: 0 to 1)
- tempo: beats per minute
- loudness: the higher value the louder the song

## Data transformation

```python
# Column duration_ms is an integer type depicting the song duration in milliseconds.
# Converting this into a more readable format

# Calculate minutes, seconds, and milliseconds from the duration_ms column
# {x // 60000:02} - 1 minute = 60,000 milliseconds
# {(x % 60000) // 1000:02} - Modulus to get remainder after minutes, then divide by 1000 to get seconds
# {x % 1000:03} - Remaining milliseconds
df["duration_reformatted"] = df["duration_ms"].apply(
    lambda x: f"{x // 60000:02}:{(x % 60000) // 1000:02}.{x % 1000:03}"
)
df.head()
```

```python
df["duration_reformatted"].info()
```

Resulting data type of new duration column is string. However, this won't be used for mathematical calculations, but only for better readability.

```python
# Removing leading and trailing white spaces from column genre (Electro-Pop has whitespace in the string value)
df["genre"] = df["genre"].str.strip()
```

## Outliers

```python
df.describe()
```

```python
# Checking all numerical columns for outliers
def detect_outliers(df):
    numeric_cols = df.select_dtypes(
        include=["number"]
    )  # Only consider numerical columns

    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = numeric_cols.quantile(0.25)
    Q3 = numeric_cols.quantile(0.75)
    IQR = Q3 - Q1

    # Determine outliers (Return True for outliers)
    outliers = (numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))

    return outliers


outliers = detect_outliers(df)
outliers.sum()
```

Outliers exist in allmost all numerical columns.
However, the numerical values in these columns are all within the expected ranges.
Even though a song can mathematically be an outlier on 'acousticness', songs can be much less or more acoustic than the broad range.
Outliers here therefore are not removed from the dataset and further analysis.

Energy of  a song is strongly correlated with loudness - the more energetic, the louder the song. 
Energy also has a fairly strong correlation to valence (positivity). More energetic songs also have more valence. 

Danceability has the strongest correlation to valence. 

Acousticeness and instrumentalness are positively correlated. Higher acousticness tends to also be more instrumental.

Valence, or the positivity of a song, is most correlated to higher energy, danceability and loudness of a song.
